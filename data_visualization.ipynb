```python
import pandas as pd
from matplotlib import pyplot as plt

import numpy as np

import seaborn as sns
```

```python
google_ratings = pd.read_csv("google_ratings.csv")

google_ratings = google_ratings[['business_id','rating']]
```

```python
yelp = pd.read_csv("test2.csv")
```

```python
yelp_google = pd.merge(left=yelp,right=google_ratings,how='inner',on='business_id')
```

```python
yelp_google['difference'] = yelp_google['rating'] - yelp_google['stars']
```

```python
yelp_google_1 = yelp_google[['business_id','difference','review_count']]
```

```python
plt.scatter(yelp_google['difference'],yelp_google['review_count'])
plt.title("Relationship between Rating Differences and Review Count")
plt.xlabel("Ratings Differences")
plt.ylabel("Review Count")
```

```python
sns.distplot(yelp_google_1['difference'],axlabel='Rating differences')
```

```python
sns.kdeplot(yelp_google_1['difference'], shade=True)
```
